```python
#[Course Name ::: Machine Learning A-Z ]
#[Part 2 - Random Forest Tree ]
#[Downloading the dataset .... www.superdatascience.com/machine-learning]

#[First Step - Importing the libraries]
#[3 essential libraries --- Numpy,matplotlib,pandas]
import numpy as np;
import matplotlib.pyplot as plt;
import pandas as pd;
print('All the above libraries are imported')
```

```python
# Importing the dataset . The file is under the same directory as the conda root.
dataset = pd.read_csv('Datasets//Part 2 - Regression//Section 9 - Random Forest Regression//Random_Forest_Regression//Random_Forest_Regression//Position_Salaries.csv');
```

```python
level = dataset.iloc[:, 1:2].values;
salary = dataset.iloc[:, 2].values;
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(level, salary)
y_pred = regressor.predict(6.5)
# The above plot is obviously wrong because for CART, 
#Plotting in higher resolution.
# The below constructs  splits the range 1-10 into 1000 points
level_grid = np.arange(min(level), max(level), 0.0001)
level_grid = level_grid.reshape(len(level_grid), 1)
plt.scatter(level, salary, color="red")
plt.plot(level_grid, regressor.predict(level_grid), color = "blue")
plt.title("Decision tree result in high resolution")
plt.xlabel("Level in high resolution")
plt.ylabel("salary")
plt.show()
```
